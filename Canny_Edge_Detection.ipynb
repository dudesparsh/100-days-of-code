```
# Importing the libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = cv2.imread('/content/5.jpg')
```

```
plt.imshow(img)
```

```
# Function for displaying image in specified size and color correction
def display(img, cmap='None'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(img2)
```

```
display(img)
```

```
# Applying Canny edge detection
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
```

```
display(edges)
```

```
# Finetuning the edges
edges = cv2.Canny(image=img, threshold1=230, threshold2=230)
```

```
display(edges)
```

```
# Choosing thresholds
med_val = np.median(img)
```

```
# Setting lower bound to be 70% of max value or positive
low = int(max(0,0.7*med_val))
```

```
# Similarly setting upper bound
upp = int(min(255, 1.3*med_val))
```

```
edges = cv2.Canny(image= img, threshold1=low, threshold2=upp)
```

```
display(edges)
```

```
# Clearly, this is not a good choice, trying blurring the images first
blur_img = cv2.blur(img, ksize=(2,2))
```

```
edges = cv2.Canny(blur_img, threshold1=low, threshold2=upp)
```

```
display(edges)
```

```
low, upp
```

```
# Best coming out for following parameters
edges = cv2.Canny(image=blur_img, threshold1=50, threshold2=200)
edges = cv2.cvtColor(edges, cv2.COLOR_BGR2RGB)
```

```
display(edges, cmap='gray')
```
