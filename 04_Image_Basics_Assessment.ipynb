#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
new_img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#Since triangle is not predefined geometry in openCV, hence we will be using
#Polylines feature

vertices = np.array([[250,700],[600,700],[425, 400]], np.int32)
# reshaping the arraw to correct dimensions
pts = vertices.reshape((-1,1,2))
pts
```

```python
cv2.polylines(new_img2,[pts], isClosed=True, color=(0,0,255), thickness=10)
plt.imshow(new_img2)
```

### BONUS TASK. Can you figure our how to fill in this triangle? It requires a different function that we didn't show in the lecture! See if you can use google search to find it.

[CLICK ME FOR A DIRECT LINK TO THE HINT](https://docs.opencv.org/3.0-beta/modules/imgproc/doc/drawing_functions.html#fillpoly)

```python
new_img3 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.fillPoly(new_img3,[pts], color=(0,20,255))
plt.imshow(new_img3)
```
